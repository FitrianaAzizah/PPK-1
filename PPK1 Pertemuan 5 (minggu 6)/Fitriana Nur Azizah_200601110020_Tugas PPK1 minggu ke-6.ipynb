### Menampilkan Judul (Title) dari Grafik

```python
import matplotlib.pyplot as plt
fig = plt.figure()
x = [10, 25, 30, 35]
y = [10, 25, 30, 35]
ax1 = plt.axes()
ax1.plot(x,y)
plt.title('Grafik garis(x)')
plt.show()
```

### Memberikan Label pada Sumbu x dan y

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,2,0.025)*np.pi
y = np.sin(x)
plt.plot(x,y,'y', label=r'$ \mathcal{x = sumbu(x)} $')

y2 = np.cos(x)
plt.plot(x,y2,'g', label=r'$ \mathcal{y = sumbu(y)} $')

plt.legend(loc='lower left')
plt.show()
```

### Menambahkan Teks pada Grafik

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,2,0.5)*np.pi
y = np.sin(x)
plt.plot(x,y, label=r'$ \mathcal{x = sumbu(x)} $')
plt.xlabel('Waktu (s)')

y2 = np.cos(x)
plt.plot(x,y2,'b', label=r'$ \mathcal{y = sumbu(y)} $')
plt.ylabel('Volume (V)')

plt.legend(loc='lower left')
plt.show()
```

1. Buatlah program yang meminta masukan user sebuah bilangan bulat N dimana (N > 0). Program kemudian menampilkan penjumlahan N bilangan genap positif pertama (bilangan genap  0).

```python
def pengulangan():
    N = int(input('Masukkan Nilai N: '))
    count = 0
    sum = 0
    for i in range (0,N*2,2):
        if i % 2 == 0:
            count = count + 1
            sum = sum + i
            print('Bilangan Genap Positif Ke -',count, ' :',i)
    print()       
    print('Jumlah bilangan genap positif dari',N,'bilangan diatas adalah',sum)
    
    print ('')
    print ('ingin coba lagi ? (ya/tidak)')
    x=input()
    if x=='ya':
        return pengulangan()
    if x=='tidak':
        print('Selesai!')
        
pengulangan()
```

2. Buatlah sebuah program yang meminta masukan user sebuah bilangan bulat N dimana (N > 0). Kemudian, program menampilkan penjumlahan N bilangan kuadrat pertama. Bilangan kuadrat adalah = 1, 4, 9, 16, 25, 36, ...., N2.

```python
def pengulangan():
    N = int(input('Masukkan Nilai N: '))
    count = 0
    sum = 0
    for i in range(1,N+1):
        if i > 0:
            count = count + 1
            sum = sum + i*i
            print('Bilangan Kuadrat Ke -',count, ' :',i*i)
    print()       
    print('Jumlah bilangan Kuadrat dari',N,'bilangan diatas adalah',sum)
    
    print ('')
    print ('ingin coba lagi ? (ya/tidak)')
    x=input()
    if x=='ya':
        return pengulangan()
    if x=='tidak':
        print('Selesai!')
        
pengulangan()
```
